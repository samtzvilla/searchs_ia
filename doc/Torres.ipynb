```python
import sys,os
baile_path=os.path.abspath(os.path.join("..","src"))
if not baile_path in sys.path:
    sys.path.append(baile_path)
import SimpleSearch as ss
```

```python
inicio=ss.node(((1,2,3,4,5,6,7),(), ()), # estado
       op='start', parent=None, cost=0)
final=ss.node(((),(), (1,2,3,4,5,6,7)))
```

```python
def meta(*estados):
    inicio=estados[0]
    final=estados[1]
    return inicio.state==final.state
```

```python
def mover(nodo):
    T=nodo.state
    sucesores=[]
    for o,to in enumerate(T):
        for d, td in enumerate(T):
            if o!=d and  len(to)>0:
                if len(td)==0 or to[0]<td[0]:
                    n=[list(t) for t in T]
                    n[o]=to[1:]
                    n[d]=to[:1]+td
                    s=tuple([tuple(t) for t in n])
                    nn=ss.node(s, depth=nodo.depth+1, 
                               op=f'mover de {o} a {d}', parent=nodo)
                    sucesores.append(nn)
    return sucesores
```

```python
bfs=ss.BlindSearch(inicio,mover,meta, goal_state=final, strategy='bfs')
```

```python
r=bfs.find()
```
